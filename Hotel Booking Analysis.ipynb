# **Project Summary -**

## Introduction

The hospitality industry is one of the most competitive markets, where understanding customer behavior and booking patterns can significantly improve business performance. This project focuses on analyzing hotel booking data to uncover meaningful insights that can help hoteliers optimize occupancy rates, reduce cancellations, improve guest satisfaction, and increase revenue. The main objective was to explore booking trends, cancellation behavior, and seasonal effects on demand, while also identifying key factors influencing customer decisions.

## Data and Methodology

The analysis was performed on a real-world dataset containing approximately 120,000 booking records from a city hotel and a resort hotel. The dataset included 32 features, such as ( hotel, is_canceled, lead_time, arrival_date_year, arrival_date_month, stays_in_weekend_nights, stays_in_week_nights, adults, children, babies, market_segment, distribution_channel, is_repeated_guest, previous_cancellations, and adr ).

The data was cleaned and preprocessed to handle missing values, duplicates, and inconsistent entries. Exploratory Data Analysis (EDA) was performed using Python libraries such as Pandas, Matplotlib, and Seaborn. Various visualizations such as bar charts, line plots, histograms, heatmaps, and pairplots were used to clearly identify booking patterns, seasonal demand, cancellation trends, and correlations among variables. These visual representations allowed for a deeper and more intuitive understanding of the dataset.

## Key Findings and Insights

The analysis revealed several interesting insights:

1. Booking Distribution: Bar plots showed that city hotels received more bookings compared to resort hotels, highlighting higher demand in urban areas. However, visualizations of stay duration confirmed that resort hotels had longer average stays, suggesting leisure travelers prefer resorts for extended vacations.

2. Seasonal Trends: Line plots and time-series visualizations indicated clear seasonal patterns, with peak bookings during summer months and holiday seasons. Off-peak seasons showed lower occupancy rates, suggesting the need for targeted discounts or promotional offers during these periods.

3. Cancellation Patterns: Heatmaps and scatterplots highlighted a strong link between longer lead times (the time between booking and arrival) and higher cancellation rates, especially for city hotels. This indicates that guests who book far in advance are more likely to cancel their plans.

4. Market Segment Insights: Visualizations of booking channels showed that most bookings came from online travel agents, underlining the importance of digital platforms in the hospitality industry. However, these channels also had higher cancellation rates compared to direct bookings, suggesting hotels could benefit from strengthening direct booking channels through loyalty programs or special incentives.

5. Guest Demographics: Plots of customer type distributions revealed that families and groups tended to book resort hotels, while solo and business travelers preferred city hotels. Additionally, visualizations of cancellations by country indicated that guests from certain regions had a higher tendency to cancel, which could guide region-specific strategies.

6. Room Preferences: Visualizations comparing requested vs. assigned room types showed mismatches that were linked with customer dissatisfaction and higher cancellations. Ensuring proper room allocation can therefore enhance guest experience and reduce booking cancellations.

## Conclusion and Recommendations

This project highlights that hotel bookings are influenced by seasonality, booking channels, lead time, and customer type. The high cancellation rates, particularly from online travel agents and long-lead bookings, pose a challenge for revenue management.

To address these issues, hotels can adopt the following strategies:

  * Cancellation rates rise as the price does. In order to prevent cancellations of reservations, hotels could work on their pricing strategies and try to lower the rates of specific hotels based on locations. They can also provide some discounts to the consumers.

  * As the ratio of the cancellation and not cancellation of the resort hotel is higher in the resort hotel than the city hotels. So the hotels should provide a reasonable discount on the room prices on weekends or on holidays

  * In the month of January, hotels can start campaigns or marketing with a reasonable amount to increase their revenue as the cancellation is the highest in this month.

  * They can also increase the quality of their hotels and their services mainly in Portugal to reduce the cancellation rate.

By leveraging data-driven strategies, hotels can strengthen their market position and improve long-term profitability.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
hotel_booking_df=dataset.copy()
````

````python
# removing duplicate rows
hotel_booking_df.drop_duplicates(inplace=True)
hotel_booking_df.reset_index(drop=True,inplace=True)
unique_rows=hotel_booking_df.shape[0]
print(f"No. of unique rows are {unique_rows}")
````

````python
# Write your code to make your dataset analysis ready.
# Count the null value for each columns in descending order.
hotel_booking_df.isnull().sum().sort_values(ascending=False).head()
````

````python
# Finding the percentage of null values present in each columns
hotel_null_percentage=((hotel_booking_df.isnull().sum()/hotel_booking_df.shape[0])*100).sort_values(ascending=False)
hotel_null_percentage.head()
````

````python
# Percentage of Null values in 'company' column is greater than 90 percent, so it would be nice to drop this column from hotel_booking_df dataframe
hotel_booking_df.drop(['company'],axis=1,inplace=True)
hotel_booking_df.head()
````

````python
# filling the missing value of country columns
hotel_booking_df['country'].fillna('others',inplace=True)
print(f"No. of NULL values in country column is {hotel_booking_df['country'].isnull().sum()}.")
````

````python
# finding the skewness value of columns 'agent' and 'children' before filling the missing values.
hotel_booking_df[['agent','children']].skew().rename('skew value')
````

````python
# filling the missing value of 'agent' and 'children' column by there median values, because there skewness values are greater than 0.5
hotel_booking_df.fillna({'agent':hotel_booking_df['agent'].median(),'children':hotel_booking_df['children'].median()},inplace=True)
print(f"No. of NULL values in agent column is {hotel_booking_df['agent'].isnull().sum()}.")
print(f"No. of NULL values in children column is {hotel_booking_df['children'].isnull().sum()}.")
````

````python
# Now checking the NULL values for each columns.
hotel_booking_df.isnull().sum().sort_values(ascending=False).head()
````

````python
# Checking the shape of hotel_booking_df dataframe after droping the duplicate values
hotel_booking_df.shape
````

````python
# Droping the rows where columns ( 'adults', 'children', and 'babies') is equal to zero. becasue it simply means no bookings were made.
hotel_booking_df=hotel_booking_df[hotel_booking_df['adults']+hotel_booking_df['children']+hotel_booking_df['babies']!=0]
````

````python
# replacing the month name with month number
hotel_booking_df['arrival_date_month']=hotel_booking_df['arrival_date_month'].replace({'January':'01','February':'02','March':'03','April':'04','May':'05','June':'06','July':'07','August':'08','September':'09','October':'10','November':'11','December':'12'})
````

````python
# Creating a column named arrival date by concat and converting arrival_date_year, arrival_date_month and arrival_date_day_of_month.
hotel_booking_df['arrival_date_str'] = hotel_booking_df['arrival_date_year'].astype(str) + '-' + hotel_booking_df['arrival_date_month'].astype(str) + '-' + hotel_booking_df['arrival_date_day_of_month'].astype(str)
hotel_booking_df['arrival_date'] = pd.to_datetime(hotel_booking_df['arrival_date_str'])
hotel_booking_df['arrival_date'].head()
````

````python
hotel_booking_df.shape
````

````python
# Data type of column 'reservation_status_date' is object, therefore I convert it to date type.
hotel_booking_df['reservation_status_date']= pd.to_datetime(hotel_booking_df['reservation_status_date'])
````

````python
# Now checking the data type of column 'reservation_status_date'
print(hotel_booking_df['reservation_status_date'].dtype)
````

````python
# Correlation Heatmap visualization code

# Selecting the columns which data type is 'integer' and 'float'
numeric_df = hotel_booking_df.select_dtypes(include=['float64', 'int64'])

# Calculate correlation matrix
corr_matrix = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True,cmap='Reds', linecolor='black',fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
````

##### 1. Why did you pick the specific chart?

This chart, called a correlation heatmap, is the best way to see the relationship between all the numerical variables in a dataset at a single glance. The color-coding makes it incredibly easy to quickly spot which factors have a strong or weak connection to each other, which would be hard to see in a simple table of numbers.

##### 2. What is/are the insight(s) found from the chart?

This heatmap reveals how different factors are connected to booking cancellations and each other.

It shows that there's a slight tendency for cancellations to increase as the **lead_time** (the gap between booking and arrival) gets longer.

Interestingly, guests who have **special_requests** or need **car_parking_spaces** are slightly less likely to cancel their bookings.

We also see some common-sense relationships, like a moderate link between the number of weekend nights and week nights a person stays.

However, the main takeaway is that most factors do not have a strong, direct impact on each other.
